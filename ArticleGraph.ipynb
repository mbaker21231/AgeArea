# Simulating Stochastic Differential Equations

```python
sigma = .51
r     = 2
tau   = .5
K     = 3

thresh = 2.5
drop   = .4

dt    = .005
T     = 1
n     = int(T/dt)
t     = np.linspace(0., T, n)

sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt    = np.sqrt(dt)

random.seed(41)

x = np.zeros(n)
x[0] = 1
breached = 0

for i in range(n - 1):
    x[i + 1] = x[i] + dt * r * x[i]*(1 - x[i] / K) + x[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (x[i + 1] >= thresh) & (breached == 0):
        xtime = i
        x[i + 1] = thresh/2
        K = drop
        breached = 1

y    = np.zeros(n)
y[0] = 1.25        
K    = 3
breached = 0
random.seed(30)

for i in range(n - 1):
    y[i + 1] = y[i] + dt * r * y[i]*(1 - y[i] / K) + y[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (y[i + 1] >= thresh) & (breached == 0):
        ytime = i
        y[i + 1] = thresh/2
        K = drop    
        breached = 1

z    = np.zeros(n)
z[0] = 1.25        
K    = 3
breached = 0
random.seed(6698)

for i in range(n - 1):
    z[i + 1] = z[i] + dt * r * z[i]*(1 - z[i] / K) + z[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (z[i + 1] >= thresh) & (breached == 0):
        ztime = i
        z[i + 1] = thresh/2
        K = drop    
        breached = 1

w    = np.zeros(n)
w[0] = 1.25        
K    = 3
breached = 0
random.seed(61)      
        
for i in range(n - 1):
    w[i + 1] = w[i] + dt * r * w[i]*(1 - w[i] / K) + w[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (w[i + 1] >= thresh) & (breached == 0):
        w[i + 1] = thresh/2
        K = drop    
        breached = 1    

        
# Data for exponential density:

lam =4.5
xpts = np.arange(0,n)/n
ppts = lam*np.exp(-lam*xpts)
```

```python
sigma = .51
r     = 2
tau   = .5
K     = 3

thresh = 2.5
drop   = .4

dt    = .0015
T     = 1
n     = int(T/dt)
t     = np.linspace(0., T, n)

sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt    = np.sqrt(dt)

random.seed(41)

x = np.zeros(n)
x[0] = 1
breached = 0

for i in range(n - 1):
    x[i + 1] = x[i] + dt * r * x[i]*(1 - x[i] / K) + x[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (x[i + 1] >= thresh) & (breached == 0):
        xtime = i
        x[i + 1] = thresh/2
        K = drop
        breached = 1

y    = np.zeros(n)
y[0] = 1.25        
K    = 3
breached = 0
random.seed(30)

for i in range(n - 1):
    y[i + 1] = y[i] + dt * r * y[i]*(1 - y[i] / K) + y[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (y[i + 1] >= thresh) & (breached == 0):
        ytime = i
        y[i + 1] = thresh/2
        K = drop    
        breached = 1

z    = np.zeros(n)
z[0] = 1.25        
K    = 3
breached = 0
random.seed(6698)

for i in range(n - 1):
    z[i + 1] = z[i] + dt * r * z[i]*(1 - z[i] / K) + z[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (z[i + 1] >= thresh) & (breached == 0):
        ztime = i
        z[i + 1] = thresh/2
        K = drop    
        breached = 1

w    = np.zeros(n)
w[0] = 1.25        
K    = 3
breached = 0
random.seed(61)      
        
for i in range(n - 1):
    w[i + 1] = w[i] + dt * r * w[i]*(1 - w[i] / K) + w[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (w[i + 1] >= thresh) & (breached == 0):
        w[i + 1] = thresh/2
        K = drop    
        breached = 1    

        
# Data for exponential density:

lam =4.5
xpts = np.arange(0,n)/n
ppts = lam*np.exp(-lam*xpts)

# Trim down the graphs

y  = y[:n-xtime]
yt = np.arange(0, n-xtime) * dt 

z  = z[:n-xtime-ytime]
zt = np.arange(0, n-xtime-ytime) * dt

w  = w[:n-xtime-ytime-ztime]
wt  = np.arange(0,n-xtime-ytime-ztime)*dt
```

Here, let's see if we can decide how we want the micro foundations to work. The trick is to get things set up so that:
1. Societies have only one active chain at a time.
2. The chains can go dormant upon "rectification."
3. Standard overlapping generations model. 

So, utility, income, children, and population next period is:

$$
p_{t+\Delta}=p_ty_t=p_tf(p_t,e_t(\Delta);\theta_t)
$$

We suppose that this can be expanded as follows:

$$
p_{t+\Delta} \approx p_t(1+r -rp_t/K+\sigma (\epsilon_{t+\Delta}-\epsilon_t))
$$

If we take $\epsilon$ to be governed by infitesmal brownian motion, we arrive at the limit of the above:

$$
dp = r\left(1-\frac{p}{K}\right)+\sigma p dz
$$

Where $dz$ is a standard brownian motion.

A couple of notes: 
1. Steady-state population is where $\overline{p}=K$. 
2. Steady-state utility is $1+r-\frac{r\overline{p}}{K}$. This simplifies to $1$ when $p$ is at or around carrying capacity.
3. We require that the drift parameter be greater than the variance divided by two: I think in our case this is $r>\sigma^2/2$.

A couple of questions: 
1. Can the threshold be at K?
2. Where does population typically hang around?

If the threshold in the new location - call it $B$ is hit, the carrying capacity changes to a $K-D$ for the current generation. Once the "time of troubles" pass (one generation) a new $K,B$ combo are drawn at the location. 

Just prior to the barrier being hit, utility is just about:
$$
u_o=1+r\left(1-\frac{B}{K}\right)
$$
in the current location. In fact, this is a lower bound on utility. So we will get no moves so long as this remains greater than:
$$
u_n=1+r\left(1-\frac{M}{K}\right)-c
$$

Let's imagine that the barrier is the same as the carrying capacity, and that $c$ is equal to one. Then we have, when the population gets to $B=K$:

$$
u_o=1, \quad u_n =r \left(1-\frac{M}{K}\right)
$$

So, if
