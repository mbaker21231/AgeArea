# Simulating Stochastic Differential Equations

```python
sigma = .51
r     = 2
tau   = .5
K     = 3

thresh = 2.5
drop   = .4

dt    = .005
T     = 1
n     = int(T/dt)
t     = np.linspace(0., T, n)

sigma_bis = sigma * np.sqrt(2. / tau)
sqrtdt    = np.sqrt(dt)

random.seed(41)

x = np.zeros(n)
x[0] = 1
breached = 0

for i in range(n - 1):
    x[i + 1] = x[i] + dt * r * x[i]*(1 - x[i] / K) + x[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (x[i + 1] >= thresh) & (breached == 0):
        xtime = i
        x[i + 1] = thresh/2
        K = drop
        breached = 1

y    = np.zeros(n)
y[0] = 1.25        
K    = 3
breached = 0
random.seed(30)

for i in range(n - 1):
    y[i + 1] = y[i] + dt * r * y[i]*(1 - y[i] / K) + y[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (y[i + 1] >= thresh) & (breached == 0):
        ytime = i
        y[i + 1] = thresh/2
        K = drop    
        breached = 1

z    = np.zeros(n)
z[0] = 1.25        
K    = 3
breached = 0
random.seed(6698)

for i in range(n - 1):
    z[i + 1] = z[i] + dt * r * z[i]*(1 - z[i] / K) + z[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (z[i + 1] >= thresh) & (breached == 0):
        ztime = i
        z[i + 1] = thresh/2
        K = drop    
        breached = 1

w    = np.zeros(n)
w[0] = 1.25        
K    = 3
breached = 0
random.seed(61)      
        
for i in range(n - 1):
    w[i + 1] = w[i] + dt * r * w[i]*(1 - w[i] / K) + w[i]**2 * sigma_bis * sqrtdt * random.normalvariate(0, 1)
    if (w[i + 1] >= thresh) & (breached == 0):
        w[i + 1] = thresh/2
        K = drop    
        breached = 1    

        
# Data for exponential density:

lam =4.5
xpts = np.arange(0,n)/n
ppts = lam*np.exp(-lam*xpts)
```

```python
fig = plt.figure(figsize=(8,6))

gs = gridspec.GridSpec(5, 2)

ax1 = plt.subplot(gs[0,0])
ax1.plot(t, x, lw=2)
ax1.plot([0,1],[thresh,thresh], linestyle='dashed', color='black')
ax1.axis('off')
an1 = ax1.annotate('A', xy=(1.05,1), size=15)
ax1.set_xlim([0, 1.2])
ax1.set_ylim([0,thresh + .5])

ax2 = plt.subplot(gs[1,0])
ax2.plot(t, y, lw=2, color='red')
ax2.plot([0,1],[thresh,thresh], linestyle='dashed', color='black')
ax2.axis('off')
an2 = ax2.annotate('B', xy=(1.05,1), size=15)
ax2.set_xlim([0, 1.2])
ax2.set_ylim([0, thresh + .5])

ax3 = plt.subplot(gs[2,0])
ax3.plot(t, z, lw=2, color='green')
ax3.plot([0,1],[thresh,thresh], linestyle='dashed', color='black')
ax3.axis('off')
an3 = ax3.annotate('C', xy=(1.05,1), size=15)
ax3.set_xlim([0, 1.2])
ax3.set_ylim([0, thresh + .5])

ax4 = plt.subplot(gs[3,0])
ax4.plot(t, w, lw=2, color='orange')
ax4.plot([0,1],[thresh,thresh], linestyle='dashed', color='black')
ax4.axis('off')
an4 = ax4.annotate('D', xy=(1.05,1), size=15)
ax4.set_xlim([0, 1.2])
ax4.set_ylim([0, thresh + .5])

# Circles for the hit points:

ax1.plot(xtime*dt, thresh, marker='o', markersize=10, markeredgecolor='blue', markerfacecolor='none')
ax2.plot(ytime*dt, thresh, marker='o', markersize=10, markeredgecolor='blue', markerfacecolor='none')
ax3.plot(ztime*dt, thresh, marker='o', markersize=10, markeredgecolor='blue', markerfacecolor='none')

# Exp. Density

ax5 = plt.subplot(gs[4,0])
ax5.plot(xpts, ppts, lw=2, color='black')
ax5.axis('off')
an5 = ax5.annotate('Exp. Density', xy=(1.1,1))
ax5.set_xlim([.10, 1.3])
ax5.set_ylim([0, thresh + .5])

ax1.axvline(x=xtime*dt, ymin=-4.6, ymax=.8, c='black', lw=1, linestyle='dashed', 
        zorder=0, clip_on=False)

ax2.axvline(x=ytime*dt, ymin=-3.55, ymax=.8, c='black', lw=1, linestyle='dashed', 
        zorder=0, clip_on=False)

ax3.axvline(x=ztime*dt, ymin=-1.9, ymax=.8, c='black', lw=1, linestyle='dashed', 
        zorder=0, clip_on=False)

ax6 = plt.subplot(gs[1:-1, 1])
ax6.set_ylim((0, 10))
ax6.set_xlim((0, 7))
ax6.axis('off')


lines = [[[0,1.45], [5,5]],
         [[1.5,6], [5,9]],
         [[1.5,6], [5,1]],
         [[4.75,6], [7.75,6.5]],
         [[5.4,6.1], [7.1,7.6]]]

for line in lines:
    ax6.plot(line[0], line[1], linewidth=3, color='blue', zorder=0) 
    
an61 = ax6.annotate('A', xy=(6.5,1), size=14)  
an62 = ax6.annotate('B', xy=(6.5,9), size=14)
an63 = ax6.annotate('C', xy=(6.5,6.5), size=14)
an64 = ax6.annotate('D', xy=(6.5,7.5), size=14)
```

# Some algebra for the paper

Here, let's see if we can decide how we want the micro foundations to work. The trick is to get things set up so that:
1. Societies have only one active chain at a time.
2. The chains can go dormant upon "rectification."
3. Standard overlapping generations model. 

So, utility, income, children, and population next period is:

$$
p_{t+\Delta}=p_ty_t=p_tf(p_t,e_t(\Delta);\theta_t)
$$

We suppose that this can be expanded as follows:

$$
p_{t+\Delta} \approx p_t(1+r -rp_t/K+\sigma (\epsilon_{t+\Delta}-\epsilon_t))
$$

If we take $\epsilon$ to be governed by infitesmal brownian motion, we arrive at the limit of the above:

$$
dp = r\left(1-\frac{p}{K}\right)+\sigma p dz
$$

Where $dz$ is a standard brownian motion.

A couple of notes: 
1. Steady-state population is where $\overline{p}=K$. 
2. Steady-state utility is $1+r-\frac{r\overline{p}}{K}$. This simplifies to $1$ when $p$ is at or around carrying capacity.
3. We require that the drift parameter be greater than the variance divided by two: I think in our case this is $r>\sigma^2/2$.

A couple of questions: 
1. Can the threshold be at K?
2. Where does population typically hang around? 

Let's first do the symbolic work and then come back and get this stuff figured out.

```python
from sympy import *
from sympy.interactive import init_printing

init_printing()
```

```python
p, K, D, r, B, c, m, M = symbols('p K D r B c m M') 
```

If the threshold in the new location - call it $B$ is hit, the carrying capacity changes to a $K-D$ for the current generation. Once the "time of troubles" pass (one generation) a new $K,B$ combo are drawn at the location. 

Just prior to the barrier being hit, utility is just about:
$$
u_o=1+r\left(1-\frac{B}{K}\right)
$$
in the current location. In fact, this is a lower bound on utility. So we will get no moves so long as this remains greater than:
$$
u_n=1+r\left(1-\frac{M}{K}\right)-c
$$

Let's imagine that the barrier is the same as the carrying capacity, and that $c$ is equal to one. Then we have, when the population gets to $B=K$:

$$
u_o=1, \quad u_n =r \left(1-\frac{M}{K}\right)
$$

So, if 

$$
r\left(1-\frac{M}{K}\right)<1
$$

No one moves unless there is a calamity. Once the barrier is hit, we have a condition of the form:

$$
1+r\left(1-\frac{B-m}{K-D}\right)=r\left(1-\frac{m}{K}\right)
$$
